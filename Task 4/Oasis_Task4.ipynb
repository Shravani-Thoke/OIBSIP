```python
import pandas as pd
```

```python
data=pd.read_csv('/content/spam.csv',encoding='latin-1')
data.head()
```

```python
data.info()
```

```python
data=data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
x=data['v2']
y=data['v1']
```

```python
y = y.map({'ham': 0, 'spam': 1})
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(stop_words='english')
x_vectorized=vectorizer.fit_transform(x)
```

```python
x_vectorized.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_vectorized,y,test_size=0.2,random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```

```python
sample_data=["Congratulations! You have won a free lottery. Claim now."]
sample_vector=vectorizer.transform(sample_data)
pred=model.predict(sample_vector)
print("Prediction of email :","Spam" if pred[0]==1 else "Ham")
```
